```python
df.drop(['mainroad',	'guestroom',	'basement',	'hotwaterheating'], axis = 1,inplace = True)
```

```python
dummies = pd.get_dummies(df[['airconditioning', 'prefarea']])
```

```python
df_dummies = pd.concat([df,dummies],axis = 'columns')
```

```python
df_dummies
```

```python
# jo dummies col add huye hai agr ak column se dusare column ki value same hai to usme se ak ko drom krna padta hai nahi to 
#internally df colinear samjhta hai           uske sath sath jo column ko dummies banaye hai usko bhi drop krna padta hai 
df_dummies.drop(['airconditioning', 'prefarea', 'prefarea_yes'], axis='columns', inplace=True, errors='ignore')
```

```python
df_dummies
```

```python
x = df_dummies.drop("price",axis = 'columns') # price target variable hai isliye drop kiye hai
```

```python
x
```

```python
y = df_dummies.price
```

```python
y
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x,y)
```
